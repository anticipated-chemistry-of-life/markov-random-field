```python
import pandas as pd
```

```python
choice = "species"
file = f"./acol_{choice}_trace.txt"
```

```python
posterior = pd.read_csv(file, sep="\t")
```

```python
posterior
```

```python
trace_smaller_than_truth = (posterior.iloc[1:] <= posterior.iloc[0]).sum() / len(
    posterior
)
```

```python
import matplotlib.pyplot as plt

plt.hist(trace_smaller_than_truth, bins=20)
```

```python
which_branch_length = "species_branch_lengths_6"
plt.scatter(
    x=range(len(posterior[which_branch_length])),
    y=posterior[which_branch_length],
    s=4,
)
plt.axhline(
    posterior[which_branch_length].iloc[0],
    color="red",
    linestyle="--",
    label=f"true value: bin={posterior[which_branch_length].iloc[0]}",
)
plt.legend()
```

```python
which_branch_length = "species_branch_lengths_6"
plt.hist(posterior[which_branch_length], bins=30)
plt.axvline(
    posterior[which_branch_length].iloc[0],
    color="red",
    linestyle="--",
    label=f"true value: bin={posterior[which_branch_length].iloc[0]}",
)
plt.legend()
```
