```python
import pandas as pd
```

```python
choice = "species"
file = f"./acol_{choice}_statePosteriors.txt"
```

```python
posterior = pd.read_csv(file, sep="\t", index_col=0)
```

```python
posterior.T
```

```python
which_branch_len = f"{choice}_branch_lengths_5"
```

```python
true_branch_length = pd.read_csv(f"../acol_{choice}_simulated.txt", sep="\t")
```

```python
true_branch_length
```

```python
true_val = true_branch_length[true_branch_length.name == which_branch_len][
    "value"
].to_numpy()
```

```python
true_val
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=posterior.T, y=which_branch_len, x=range(len(posterior.T))
).set_title(which_branch_len)
plt.axvline(x=true_val, color="red")
```
